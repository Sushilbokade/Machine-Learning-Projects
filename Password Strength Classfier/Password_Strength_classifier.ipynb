## Importing Libraries:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Dataset:

```python
data = pd.read_csv("data.csv", error_bad_lines=False)
```

```python
sns.countplot(data['strength']);
```

## Creating array of the dataset

```python
password_tuple = np.array(data)
password_tuple
```

## Shuffling data to make it robust:

```python
import random
random.shuffle(password_tuple)
```

```python
x = [labels[0] for labels in password_tuple]
y = [labels[1] for labels in password_tuple]
```

```python
x
```

```python
y
```

## Create a custom function to split into characters of list:

```python
def word_divide_char(inputs):
    character = []
    for i in inputs:
        character.append(i)
    return character
```

```python
word_divide_char('a110804032')
```

## Import TF-IDF vectorizer to convert String data into numerical data:

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(tokenizer=word_divide_char)
```

## Apply TF-IDF vectorizer on data:

```python
X = vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_document_vector = X[0]
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
df = pd.DataFrame(first_document_vector.T.todense(), index=vectorizer.get_feature_names(), columns=['TF-IDF'])
df.sort_values(by=['TF-IDF'], ascending=False)
```

## Splitting data into train and test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

## Applying XGBoost Classifier:

```python
from xgboost import XGBClassifier
classifier = XGBClassifier()
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Making Predictions for specific data:

```python
d2 = np.array(['123@#@#34WEWE@#$23'])
pred = vectorizer.transform(d2)
classifier.predict(pred)
```

## Doing predictions on X_test data:

```python
y_pred = classifier.predict(X_test)
print(y_pred)
```

## Checking accuracy of the Model using:<li>confusion matrix</li><li>accuracy score</li>

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Report of the model:

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
